```python
import altair as alt
import pandas as pd
import json

alt.data_transformers.enable('json')

# Load data
complaints_df = pd.read_csv('CDPH_Environmental_Complaints.csv')

# Convert all values in the "COMPLAINT TYPE" column to uppercase
complaints_df['COMPLAINT TYPE'] = complaints_df['COMPLAINT TYPE'].str.upper()

# First, ensure that the 'COMPLAINT DATE' column is in datetime format
complaints_df['COMPLAINT DATE'] = pd.to_datetime(complaints_df['COMPLAINT DATE'])

# Then, create a new column 'COMPLAINT YEAR' to store the year
complaints_df['COMPLAINT YEAR'] = complaints_df['COMPLAINT DATE'].dt.year
complaints_df['COMPLAINT MONTH'] = complaints_df['COMPLAINT DATE'].dt.month_name()

print(complaints_df.head())
```

```python
# Define a selection for the line chart based on complaint type
line_selection = alt.selection_single(empty='all', on='click', fields=['COMPLAINT TYPE'])

# Define the Altair chart for the stacked line chart with the selection
line_chart = alt.Chart(complaints_df).mark_area().encode(
    x=alt.X('COMPLAINT YEAR:O', title='Year'),
    y=alt.Y('count():Q', title='Number of Complaints'),
    color=alt.Color('COMPLAINT TYPE:N', scale=alt.Scale(scheme='dark2'), title='Complaint Type'),
    tooltip=['COMPLAINT YEAR:Q', 'count():Q', 'COMPLAINT TYPE:N']
).properties(
    width=600,
    height=400,
    title='Stacked Line Chart of Complaints Over the Years'
).add_selection(
    line_selection
)

# Define the Altair chart for the grid chart with the line chart selection
grid_chart = alt.Chart(complaints_df).mark_rect().encode(
    x=alt.X('COMPLAINT YEAR:O', title='Year'),
    y=alt.Y('COMPLAINT MONTH:O', title='Month', sort=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']),
    color=alt.Color('count():Q', scale=alt.Scale(scheme='viridis')),
    tooltip=['COMPLAINT YEAR:N', 'COMPLAINT MONTH:N', 'count():Q']
).transform_filter(
    line_selection
).properties(
    width=800,
    height=300,
    title='Heatmap of Count of Complaints Across Months and Years'
)

# Combine the line chart and grid chart side by side
combined_chart = line_chart & grid_chart

# Display the combined chart
combined_chart

# with open("linked_charts/linked_stacked_line_grid/linked_stacked_line_grid_raw_files.json", "w") as f:
#     f.write(json.dumps(combined_chart.to_dict(), indent=4))

# with open("../JSON/linked_stacked_line_grid_complaints_df.json", "w") as f:
#     f.write(complaints_df.to_json(indent=4))
```
